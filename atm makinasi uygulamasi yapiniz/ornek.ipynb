```python
"""
atm makinasi uygulamasi yapiniz
kullanicidan yapmak istediği işlemleri aliniz ve buna bağli olarakta bakiye, para yatirma, para çekme 
işlemlerini yapiniz
1 bakiye sorgulama 
2 para yatirma
3 para çekme
q proğramdan çikiş
bakiye = 100 olarak aliniz

"""






print("""
----------atm makinesi uygulamasi----------

1 bakiye sorgulama

2 para yatirma 

3 para çekme

proğramdan çikmak için q ya basiniz


""")

bakiye = 1000

while True:
    işlem = input("işlem seçiniz: ")
    
    if işlem == "q" :
        print("işlemden çikiliyor...")
        break
        
    elif işlem == "1" :
        print("bakiyeniz : ", bakiye)
        
    elif işlem == "2" :
        miktar = int(input("miktar giriniz : "))
        bakiye += miktar
        print("bakiyeniz : ", bakiye)
        
    elif işlem == "3":
        miktar = int(input("miktar giriniz : "))
        
        if bakiye < miktar :
            print("çekilecek miktar bakiyenizden büyük olamaz")
            continue
        
        bakiye -= miktar
        print("kalan baiyeniz : ", bakiye)
        
        
    else:
        print("geçersiz işlem")
```
