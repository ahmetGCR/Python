```python
# geometrik şekil

"""
*
**
***
****
"""

# sola yaslamali azalan yildizli üçgen

for i in range(1,5):
    for k in range(i):
        print("*", end = " ")
    print()    


# sola yaslamali artan yildizli üçgen
for i in range(5, 0, -1):
    for i in range(i):
        print("*",end=" ")
    print()   


# kare oluşturma
satir = 5
sutun = 5
for i in range(1,satir+1):
    for k in range(1,sutun +1):
        print("*", end=" ")
    print() 



"""   
   *       1 yildiz 3 boşluk
  ***      3 yildiz 2 boşluk
 *****     5 yildiz 1 boşluk
*******    7 yildiz 0 boşluk

"""


for i in range(4):                  # 0,1,2,3
    print(" " * (3-i), end="")      # 3,2,1,0
    print("*" * (i*2+1), end="")    # 1,3,5,7          
    print() 

satir_sayisi = int(input("satir sayisini giriniz : "))

for i in range(satir_sayisi):                # i      0,1,2,3
    print(" " * ((satir_sayisi-1)-i), end="")    # boşluk 3,2,1,0
    for k in range(i*2+1):        # k      1,3,5,7
        print("*", end="")
    print()      
```
