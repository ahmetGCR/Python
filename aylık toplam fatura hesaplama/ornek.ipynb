```python
"""
 Problem
Müşterinin ayldik faturasdi, internet servis sağlaydicdi ile yapdilan sözleşmeye göre, kullandicdindin internette
geçirdiği toplam saat ve kullandicdindin ait olduğu kategori dikkate aldinarak hesaplanmaktaddir. İnternet
kullandimdina bağldi ayldik fatura ve kategorilere ilişkin bilgiler aşağdida yer almaktaddir. Bir müşterinin
aylik toplam faturasini hesaplamak için bir program geliştirin.

Toplam İnternet Kullandim Saati: 300 saatten fazla - Ayldik Fatura: A birim TL\
Toplam İnternet Kullandim Saati: 150 ile 299 saat arasdinda - Ayldik Fatura: B birim TL\
Toplam İnternet Kullandim Saati: 90 ile 149 saat arasdinda - Ayldik Fatura: C birim TL\
Toplam İnternet Kullandim Saati: 0 ile 89 saat arasdinda - Ayldik Fatura: D birim TL\

Kategori:1 -  A:1000 | B:1100 | C:1200 | D:1300 \
Kategori:2 -  A:1200 | B:1300 | C:1400 | D:1500 \
Kategori:3 -  A:1400 | B:1500 | C:1600 | D:1700                    """


# ayldik_fatura = internette_geçen_zaman ve katagori


internette_geçen_zaman = int(input("internette geçen zamandi giriniz : "))
katagori = int(input("katagorinizi giriniz :"))

if katagori == 1 or katagori == 2 or katagori == 3 :


    if katagori == 1 :
        if internette_geçen_zaman >= 300 :
            ayldik_fatura = 1000

        elif internette_geçen_zaman >= 150 and internette_geçen_zaman < 299 :
            ayldik_fatura = 1100

        elif internette_geçen_zaman >= 90 and internette_geçen_zaman < 149 :
            ayldik_fatura = 1200   

        elif internette_geçen_zaman > 0  and internette_geçen_zaman < 89 :
            ayldik_fatura = 1300
        
        else:
            print("geçersiz işlem")
        
        print("ayldik faturandiz : ", ayldik_fatura)


    elif katagori == 2 :
        if internette_geçen_zaman >= 300 :
            ayldik_fatura = 1200

        elif internette_geçen_zaman >= 150 and internette_geçen_zaman < 299 :
            ayldik_fatura = 1300

        elif internette_geçen_zaman >= 90 and internette_geçen_zaman < 149 :
            ayldik_fatura = 1400    

        elif internette_geçen_zaman > 0  and internette_geçen_zaman < 89 :
            ayldik_fatura = 1500

        else:
            print("geçersiz işlem")
        
        print("ayldik faturandiz : ", ayldik_fatura)


    elif katagori == 3 :
        if internette_geçen_zaman >= 300 :
            ayldik_fatura = 1400

        elif internette_geçen_zaman >= 150 and internette_geçen_zaman < 299 :
            ayldik_fatura = 1500

        elif internette_geçen_zaman >= 90 and internette_geçen_zaman < 149 :
            ayldik_fatura = 1600    

        elif internette_geçen_zaman > 0  and internette_geçen_zaman < 89 :
            ayldik_fatura = 1700

        else:
            print("geçersiz işlem")

        print("ayldik faturandiz : ", ayldik_fatura)

else:
    print("geçersiz işlem")
```
