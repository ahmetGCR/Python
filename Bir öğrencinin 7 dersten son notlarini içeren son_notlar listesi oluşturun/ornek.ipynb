```python
"""
Bir öğrencinin 7 dersten son notlarini içeren son_notlar listesi oluşturun. Bu liste ile beraber gecti_kaldi 
isminde başka bir liste oluşturun. gecti_kaldi listesi öürencinin 7 dersten geçip geçmediğinin sonucunu 
barindirsin. Eğer öğrencinin son notu 70'ten buyükse geçti (True) aksi taktirde kaldi(False) olarak belirtilmeli.
Eğer öğrencinin
- 1. Durum: Eğer öğrenci tüm derslerinden başarili ise bir üst sinifa geçer.
- 2. Durum:  Eğer öğrencinin sadece bir tane başarisiz dersi varsa yine üst sinifa geçebilir.
- 3. Durum: Eğer öğrenci sadece bir dersten dahi geçse üst sinifa geçebilir.

"""
# 1.Durum: Eğer öğrenci tüm derslerinden başarili ile bir üst sinifa geçer.
son_notlar = [70, 70, 70, 76, 84, 80, 91]
geçti_kaldi = []

for notlar in son_notlar:
    if notlar >= 70 :
        geçti_kaldi.append(True)
    else:
        geçti_kaldi.append(False)

if all(geçti_kaldi):
    print("bir üst sinifa geçtiniz.")

else:
    print("başarisiz")


print()

# 3.Durum: Eğer öğrenci sadece bir dersten dahi geçse üst sinifa geçebilir.
son_notlar = [40, 40, 40, 76, 60, 30, 20]
gecti_kaldi = []

if any(son_notlar):
    print("bir üst sinifa geçtiniz")
else:
    print("başarisiz")

print()

# 2.Durum: Eğer öğrencinin sadece bir tane başarisiz dersi varsa yine üst sinifa geçebilir.
son_notlar = [66, 70, 77, 76, 84, 80, 91]
gecti_kaldi = []

for notlar in son_notlar:
    gecti_kaldi.append(notlar >= 70)

if gecti_kaldi.count(False) < 2:
    print("bir üst sinifa geçtiniz")
else:
    print("başarisiz")
```
