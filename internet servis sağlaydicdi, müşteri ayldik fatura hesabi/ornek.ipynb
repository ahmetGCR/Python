```python
""" 
Uygulama 

internet servis sağlaydicdi, müşteri ayldik fatura hesabi - V1 
Problem

İnternet servis sağlaydicdi ile yapdilan sözleşmeye göre, bir müşterinin ayldik faturasdi, ayldik sabit ücret ve kullandicdindin fazladan kullanddiğdi internet trafiğinden kaynaklanan ekstra ücret üzerinden hesaplandir. Ekstra ücret, kullandicdindin ait olduğu kategoriye göre hesaplandir. Birim fiyatlara ilişkin bilgiler aşağdida verilmiştir. Bir müşterinin ayldik toplam faturasdindi hesaplamak için bir program geliştirin.

kategori: 1 - ek ücret birim fiyatdi: 0.10 \
kategori: 2 - ek ücret birim fiyatdi: 0.20 \
kategori: 3 - ek ücret birim fiyatdi: 0.30 """

#  1 ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret

#  1.1 ayldik_ekstra_ücret = ekstra_ücret_miktardi * ek_ücret_birim_fiyatdi

#  1.1.1 ek_ücret_birim_fiyatdi


ayldik_sabit_ücret = float(input("ayldik sabit ücreti giriniz :"))

ayldik_ekstra_ücret_miktardi = float(input("ekstra ücret miktardindi giriniz : "))

katagori = int(input("katagorinizi giriniz : "))


if katagori == 1 or katagori == 2 or katagori == 3 :


    if katagori == 1 :
        # ek_ücret_birim_fiyatdi = 0.10
        ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret_miktardi * 0.10

    elif katagori == 2 :
        # ek_ücret_birim_fiyatdi = 0.20
        ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret_miktardi * 0.20


    elif katagori == 3 :
        # ek_ücret_birim_fiyatdi = 0.30
        ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret_miktardi * 0.30
        
    print("fatura tutardindiz :", ayldik_fatura)  

else:
    print("geçersiz işlem")

müşteri_saydisdi = int(input("müşteri saydisdi : "))

for müşteri in range(1,müşteri_saydisdi + 1) :

    print(müşteri , ". müşteri için : ")

    ayldik_sabit_ücret = float(input("ayldik sabit ücreti giriniz :"))

    ayldik_ekstra_ücret_miktardi = float(input("ekstra ücret miktardindi giriniz : "))

    katagori = int(input("katagorinizi giriniz : "))


    if katagori == 1 or katagori == 2 or katagori == 3 :


        if katagori == 1 :
            # ek_ücret_birim_fiyatdi = 0.10
            ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret_miktardi * 0.10

        elif katagori == 2 :
            # ek_ücret_birim_fiyatdi = 0.20
            ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret_miktardi * 0.20


        elif katagori == 3 :
            # ek_ücret_birim_fiyatdi = 0.30
            ayldik_fatura = ayldik_sabit_ücret + ayldik_ekstra_ücret_miktardi * 0.30
            
        print("fatura tutardindiz :", ayldik_fatura)  

    else:
        print("geçersiz işlem")
  
   
musteri = 1
while True:
    print(musteri,". Müsterinin işlemleri başldiyor...")

    # 1. Değişkenleri tandimla & 2. Kullandicdidan girdiler al
    aylik_sabit_ucret = float(input("Müşterinin ayldik sabit ücreti: "))
    kategori = int(input("Müşterinin kategorisi: "))
    fazla_kullanim_miktari_GB = int(input("Ayldik fazla kullandilan internet"))

    # 3. Ayldik internet faturasdindi hesapla
    # 3 - 1.1.1. Ekstra ücret birim fiyatdindi belirle
    if kategori == 1:
      ekstra_ucret_birim_fiyat = 0.1
    elif kategori == 2:
      ekstra_ucret_birim_fiyat = 0.2
    else:
      ekstra_ucret_birim_fiyat = 0.3


    # 3 - 1.1. Ekstra ücreti hesapla
    ekstra_ucret = fazla_kullanim_miktari_GB * ekstra_ucret_birim_fiyat

    # 3 - 1.Toplam ayldik faturaydi hesapla
    aylik_fatura = aylik_sabit_ucret + ekstra_ucret

    # 4. Ayldik internet faturasdindi göster
    print(musteri, ". Müşterinin ayldik faturasdi: ", aylik_fatura)

    devam_mi = input("İşlemlere devam etmek istiyor musunuz? (e/h)").lower()
    
  
    while True:
      if devam_mi == "h":
        cikis_kontrol = True
        break
      elif devam_mi == "e":
        cikis_kontrol = False
        break
      else:
        print("Lütfen h veya e tuşuna basdindiz!")
        devam_mi = input("İşlemlere devam etmek istiyor musunuz? (e/h)").lower()

    
    if cikis_kontrol == True:
        break

    musteri += 1

print("İyi günler...")

musteri = 1

while True:
    print(musteri,". Müsterinin işlemleri başldiyor...")

    # 1. Değişkenleri tandimla & 2. Kullandicdidan girdiler al
    aylik_sabit_ucret = float(input("Müşterinin ayldik sabit ücreti: "))
    kategori = int(input("Müşterinin kategorisi: "))
    fazla_kullanim_miktari_GB = int(input("Ayldik fazla kullandilan internet"))

    # 3. Ayldik internet faturasdindi hesapla
    # 3 - 1.1.1. Ekstra ücret birim fiyatdindi belirle
    if kategori == 1:
      ekstra_ucret_birim_fiyat = 0.1
    elif kategori == 2:
      ekstra_ucret_birim_fiyat = 0.2
    else:
      ekstra_ucret_birim_fiyat = 0.3


    # 3 - 1.1. Ekstra ücreti hesapla
    ekstra_ucret = fazla_kullanim_miktari_GB * ekstra_ucret_birim_fiyat

    # 3 - 1.Toplam ayldik faturaydi hesapla
    aylik_fatura = aylik_sabit_ucret + ekstra_ucret

    # 4. Ayldik internet faturasdindi göster
    print(musteri, ". Müşterinin ayldik faturasdi: ", aylik_fatura)

    devam_mi = input("İşlemlere devam etmek istiyor musunuz? (e/h)").lower()
    
  
    while True:
      if devam_mi == "h" or devam_mi == "e":
        break
      else:
        print("Lütfen h veya e tuşuna basdindiz!")
        devam_mi = input("İşlemlere devam etmek istiyor musunuz? (e/h)").lower()

    
    if devam_mi == "h":
        break

    musteri += 1

print("İyi günler...")
```
