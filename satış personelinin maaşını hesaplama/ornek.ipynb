```python
""" 
Bir şirketin yönetim politikasina göre, satiş elemanina sabit bir maaşi ile beraber, satiştan elde edilen toplam
kazanç ve prim oranina bağli olarak bir prim ödenir. Toplam kazancin, şirket politikasi ile belirlenen performans
düzeyini aşmasi durumunda prim orani 0.10, aksi durumda 0.05 olarak belirlenir. Şirketteki satiş elemaninin 
maaşini hesaplamak için bir program geliştirin.

"""


saat = int(input("Toplam trafik saatini girin: "))
kategori = int(input("Kategoriyi girin: "))

if saat > 0 and saat < 90:
    if kategori == 1:
        aylikFatura = 1300
    elif kategori == 2:
        aylikFatura = 1500
    elif kategori == 3:
        aylikFatura = 1700
    else:
        print('Kategori için geçersiz girdi')

elif saat >= 90 and saat < 150:
    if kategori == 1:
        aylikFatura = 1200
    elif kategori == 2:
        aylikFatura = 1400
    elif kategori == 3:
        aylikFatura = 1600
    else:
        print('Kategori için geçersiz girdi')

elif saat >= 150 and saat < 300:
    if kategori == 1:
        aylikFatura = 1100
    elif kategori == 2:
        aylikFatura = 1300
    elif kategori == 3:
        aylikFatura = 1500
    else:
        print('Kategori için geçersiz girdi')
        
else:
    if kategori == 1:
        aylikFatura = 1000
    elif kategori == 2:
        aylikFatura=1200
    elif kategori == 3:
        aylikFatura = 1400
    else:
        print('Kategori için geçersiz girdi')
    
# 4. Aylik faturayi gösterin
print('Aylik faturaniz: ', aylikFatura)
```
