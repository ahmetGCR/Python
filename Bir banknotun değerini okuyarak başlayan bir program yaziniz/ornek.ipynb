```python
"""
Bir ülkenin eski liderlerinin veya tarihsel öneme sahip diğer kişilerin resimlerinin paralarinin üzerinde 
görünmesi yaygin bir durumdur. ABD banknotlari üzerinde görünen kişiler aşağida listelenmiştir.
Bir banknotun değerini okuyarak başlayan bir program yaziniz.
Ardindan programiniz, girilen tutarin banknotunda görünen kişinin adini göstermelidir. Böyle bir banknot yoksa 
uygun bir hata mesaji görüntülenmelidir.

**Kişi ve $ değeri:**

George Washington:   $1

Thomas Jefferson:    $2

Abraham Lincoln:     $5

Alexander Hamilton:  $10

Andrew Jackson:      $20

Ulysses S. Grant:    $50

Benjamin Franklin:   $100

"""

# banknot değerini al

banknot_değeri = int(input("banknot değerinin giriniz : "))
resim = ""

if banknot_değeri == 1 :
    resim = "George Washington"

elif banknot_değeri == 2 :
    resim = "Thomas Jefferson"

elif banknot_değeri == 5 :
    resim = "Abraham Lincoln"

elif banknot_değeri == 10 :
    resim = "Alexander Hamilton"   

elif banknot_değeri == 20 :
    resim = "Andrew Jackson" 

elif banknot_değeri == 50 :
    resim = "Ulysses S. Grant"  

elif banknot_değeri == 100 :
    resim = "Benjamin Franklin"
 
if resim == "":
    print("Bu dolaşimda olan bir Amerikan banknotu değil.")   

else :
    print("Bu banknottaki kişi tanimli değil")

print(resim)
```
