```python
# örnek 50 :
"""
yil dört mevsime ayrilmiştir: ilkbahar, yaz, sonbahar ve kiş. Kullanicinin girdiği ay ve güne göre içinde 
bulunulan mevsimi gösteren programi geliştirin.
Mevsimlerin değiştiği kesin tarihler, takvimin oluşturulma biçiminden dolayi yildan yila biraz farklilik gösterse
de, kuzey yarim küre için aşağidaki verileri kullanabiliriz.

Mevsim: İlkbahar 
mevsim İlk günü: 20 Mart - Nisan - Mayis - 20 Haziran'a kadar
Mevsim: Yaz 
mevsim İlk günü: 21 Haziran - Temmuz - Ağustos - 21 Eylül'e kadar
Mevsim: Sonbahar 
mevsim İlk günü: 22 eylül - Ekim -Kasim - 20 Aralik'a kadar
Mevsim: Kiş 
mevsim İlk günü: 21 Aralik - Ocak - Şubat - 19 Mart 

"""


# input : ay ,gün

ay = input("ay'i giriniz :")
gün = input("günü giriniz : ").lower()

if ay.isalpha() and gün.isdigit():
    gün = int(gün)

    if ay == "mart" or ay == "nisan" or ay == "mayis" or ay == "haziran" :
        if ay == "nisan" or ay == "mayis" :
            mevsim = "ilkbahar"

        elif ay == "mart" and gün >= 20:
            mevsim = "ilkbahar"

        elif ay == "haziran" and gün < 21 :
            mevsim = "ilkbahar"        

    elif ay == "haziran" or ay == "temmuz" or ay == "ağustos" or ay == "eylül":
        if ay == "temmuz" or ay == "ağustos" :
            mevsim = "yaz"
        elif ay == "haziran "   and gün >= 21 :
            mevsim = "yaz"

        elif ay == "eylül" and gün < 22 :
            mevsim = "yaz"

    elif ay == "eylül" or ay == "ekim" or ay =="kasim" or ay == "aralik" :
        if ay == "ekim" or ay == "kasim":
            mevsim = "sonbahar"

        elif ay ==  "eylül" and gün >= 22 :
            mevsim = "sonbahar"

        elif ay == "aralik" and gün < 21 :
            mevsim = "sonbahar"

    elif ay == "aralik" or ay == "ocak" or ay == "şubat" or ay == "mart" :
        if ay == "ocak" or ay == "şubat":
            mevsim = "kiş"

        elif ay == "aralik" and gün >= 21 :
            mevsim = "kiş"

        elif ay == "mart" and gün < 20 :
            mevsim = "kiş"

    else:
        print("geçersiz işlem")                                    

    print(f'{gün} {ay} : {mevsim}')

else:
    print("geçerli ay ve gün girin")
```
