```python
"""
Bir dersten öğrencilerin final notlarini içeren listeyi YETKİLİDEN ALARAK oluşturunuz. Eğer öğrencinin notu 
70'ten büyük, öğrenci bu dersten geçer, aksi durumda kalir. Geçen ve kalan öğrenci listelerini ayri ayri 
list **comprehension** metodu KULLANMADAN ile oluşturunuz.

Ve oluşan listelere göre öğrencilerin isimlerini gösteriniz. Not: Öğrenci isimleri listesine ihtiyaç var. 
Final notlari ile ayni siralamada oluşturulmali.

"""

"""öğrenci_isimleri = input("isminizi girin : ")
öğrenci_notlari= int(input(f"{öğrenci_isimleri}, isimli öğrencinin notunu girin : "))
"""


print("""

çikmak için q ya basiniz    """)


öğrenci_isimleri = []
öğrenci_notlari = []

while True:
    isim = input("öğrenci ismini giriniz : ").lower()
    if isim == "q":
        print("proğramdan çikiliyor...")
        break

    notu = int(input(f"{isim} isimli öğrencinin notunu giriniz : "))
    öğrenci_isimleri.append(isim)
    öğrenci_notlari.append(notu)

geçme_notu = 70

geçenler = []
kalanlar = []

for i in range(len(öğrenci_isimleri)):
    if öğrenci_notlari[i] > geçme_notu:
        geçenler.append(öğrenci_isimleri[i])
    else:
        kalanlar.append(öğrenci_isimleri[i])

if len(geçenler) > 0:
    print("geçen öğrenciler :")
    for i in geçenler:
        print(i,end=" ")

print()

if len(kalanlar) > 0:
    print("kalan öğrenciler")

    for i in kalanlar:
        print(i, end=" ")
```
