```python
""" 
Elektrik tedarik şirketinin abonelik sözleşmesine göre, müşterinin aylik faturasi, aylik elektrik tüketimi ve 
elektriğin birim fiyatina göre hesaplanir. Elektrik sayaci okunduktan sonra sifirlanmaz. Bu nedenle, aylik 
elektrik tüketimi, son okumaya ve bir önceki okumaya göre hesaplanmalidir. Birim fiyat bilgileri aşağidaki 
tablolarda verilmiştir. Bir müşterinin aylik faturasini hesaplamak için bir program geliştirin.

Elektrik Tüketimi: 0 - 100 kWh, Birim Fiyat: 1.1

Elektrik Tüketimi: 101 - 200 kWh, Birim Fiyat: 1.3

Elektrik Tüketimi: 201 - 300 kWh, Birim Fiyat: 1.5

Elektrik Tüketimi: 301 - 400 kWh, Birim Fiyat: 1.7

Elektrik Tüketimi: 401 - 500 kWh, Birim Fiyat: 1.9

Elektrik Tüketimi: 501 kWh ve üstü, kWh, Birim Fiyat: 2.3

"""

# aylik_fatura = aylik_elektrik_tüketimi * elektrik_birim_fiyati
# yeni_okuma = float(input("sayaç : "))
# eski_okuma = float(input("sayaç : "))
# aylik_elektrik_tüketimi = yeni_okuma - eski_okuma




yeni_okuma = input("yeni okuma : ")
eski_okuma = input("eski okuma : ")


if yeni_okuma.isdigit() and eski_okuma.isdigit() :
    yeni_okuma = float(yeni_okuma)
    eski_okuma = float(eski_okuma)
    aylik_elektrik_tüketimi = abs(yeni_okuma - eski_okuma)

    if aylik_elektrik_tüketimi > 0 and aylik_elektrik_tüketimi < 100 :
        aylik_fatura = aylik_elektrik_tüketimi * 1.1

    elif aylik_elektrik_tüketimi <= 200 :
        aylik_fatura = aylik_elektrik_tüketimi * 1.3

    elif aylik_elektrik_tüketimi <= 300 :
        aylik_fatura = aylik_elektrik_tüketimi * 1.5

    elif aylik_elektrik_tüketimi <= 400 :
        aylik_fatura = aylik_elektrik_tüketimi * 1.7

    elif aylik_elektrik_tüketimi <= 500 :
        aylik_fatura = aylik_elektrik_tüketimi * 1.9

    else :
        aylik_fatura = aylik_elektrik_tüketimi * 2.3

    print("aylik faturaniz : ", aylik_fatura)        

else:
    print("lütfen rakam veya sayi giriniz")
```
